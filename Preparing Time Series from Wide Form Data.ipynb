```python
#Imports
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.dates as mdates
import matplotlib.ticker as mticks
import matplotlib.pyplot as plt
import seaborn as sns
import os
# set random seed
SEED = 321
np.random.seed(SEED)
```

```python
df = pd.read_csv('Data/City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
df.head()
```

**Filter the 4 largest cities into a new dataframe**

```python
#Create a copy of the dataframe
dff = df.copy()
```

```python
#Filter
four_largest_cities = (dff['SizeRank'] >= 0) & (dff['SizeRank'] <=3)
```

```python
dff[four_largest_cities]
```

```python
#Filter
dff = dff[four_largest_cities]
dff
```

**Melt the data to long-form and prepare it for time series analysis**

```python
dff.columns
```

```python
melt = pd.melt(dff.drop(columns=['RegionID','StateName']),
                 id_vars=['SizeRank', 'RegionName','RegionType','State','Metro','CountyName'],
                value_name='Home Value',var_name="Date")
melt
```

**Convert the melted dates into datetime datatype**

```python
melt['Date'] = pd.to_datetime(melt['Date'])
melt.info()
```

**Make the datetime column the index**

```python
melt = melt.set_index("Date")
melt
```

**Resample the dataframe as monthly frequency, grouped by City.**

```python
dff_ts = melt.groupby('RegionName').resample("M").sum()
dff_ts
```

**Plot the home values for all 4 cities**

```python
#Function to convert thousands into K
def thousands(x,pos):
    """function for use wth matplotlib FuncFormatter -  formats money in millions"""
    new_x = x/1_000
    return f"{new_x:,.0f}K"

#Create the formatter
fmt_thous = mticks.FuncFormatter(thousands)
```

```python
#Unstack each city
unstack = dff_ts.unstack(0)
```

```python
unstack
```

```python
unstack.drop('SizeRank', axis=1, inplace=True)
```

```python
ax = unstack['Home Value'].plot(figsize=(12,5))
ax.yaxis.set_major_formatter(fmt_thous)
ax.set_title('Home Values for 4 Largest Cities')
ax.set_xlabel('Date')
ax.set_ylabel('$ Dollars');
```

**Question 1: Which City had the highest Typical Home Value at the end of 2008? Which had the least?**

```python
unstack.head()
```

```python
# Get max value and index
min = unstack.idxmin()
index = unstack.loc['2008-12'].idxmax()[1]
unstack.loc[index]
```

New York had the highest typical home value at the end of 2008. Home value for New York was $510,309.

Houston had the lowest typical home value at the end of 2008. Home value for Houston was $131,283.

```python
#Plot all cities

ax = unstack['Home Value'].plot(figsize=(12,5))
ax.yaxis.set_major_formatter(fmt_thous)
ax.set_title('Home Values for 4 Largest Cities')
ax.vlines(index,0,1000000, colors='k', linestyles='dashed')
ax.yaxis.set_major_formatter(fmt_thous)
ax.set_xlabel('Date')
ax.set_ylabel('$ Dollars');
```

**Question 2: How much did the home values change from November 2008 to December 2008 (in dollars)?**

```python
#Difference between Nov to Dec 2008
difference2008 = unstack.loc['nov 2008':'dec 2008'].diff().dropna()
difference2008
```

For each city home values went down from November 2008 to December 2008.

Chicago homes lost $5,753.

Houston homes lost $964.

Los Angeles homes lost $12,611.
